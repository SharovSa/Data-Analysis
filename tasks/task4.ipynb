```python
import pandas as pd
import numpy as np
import scipy.stats as scp
```

```python
data = pd.read_csv('cps99_ps3.csv')
```

```python
ahe = data.ahe.values
ahe_mean = np.mean(ahe)
female = data.female.values
female_mean = np.mean(female)

beta1 = np.sum(female*(ahe - ahe_mean)) / np.sum(female*(female - female_mean))
beta0 = ahe_mean - beta1*female_mean

print(f"a) Коэффициент при female: {beta1}.\nЭто означает то, что женщины зарабатывают на {abs(beta1)} меньше мужчин.")
```

```python
# Если среднечасовая ставка для женщин и мужчин одинакова, то beta1 = 0.
# H0: beta1 = 0
# H1: beta1 != 0

residuals = ahe - (beta0 + beta1*female)
std_residuals = np.std(residuals, ddof=1)
t_stat = beta1 / (std_residuals / np.sqrt(np.sum((female - female_mean)**2)))
p_value = scp.t.cdf(t_stat, len(female) - 2)
print("b)")
if p_value < 0.05: 
    print("Отвергаем H0 в пользу H1 => среднечасовая ставка для женщин и мужчин различна.")
else:
    print("Принимаем H0 => среднечасовая ставка для женщин и мужчин одинакова.")
```

```python
# H0: популяционная разница в доходах равна 0
# H1: популяционная разница в доходах не равна 0
alpha = 0.05
male_ahe = data[data.female == 0].ahe
male_ahe_mean = np.mean(male_ahe)
female_ahe = data[data.female == 1].ahe
female_ahe_mean = np.mean(female_ahe)
diff = abs(male_ahe_mean - female_ahe_mean)
t_stat = diff / np.sqrt(np.var(male_ahe, ddof=1)/len(male_ahe) + np.var(female_ahe, ddof=1)/len(female_ahe))
print("c)")
if t_stat > scp.t.ppf(1-alpha/2, len(male_ahe) + len(female_ahe) - 2):
    print("Отвергаем H0 в пользу H1 => популяционная разница в доходах женщин и мужчин не равна 0.")
else:
    print("Принимаем H0 => популяционная разница в доходах для женщин и мужчин равна нулю.")

print("Результаты во всех пунктах дают один и тот же ответ")
```
